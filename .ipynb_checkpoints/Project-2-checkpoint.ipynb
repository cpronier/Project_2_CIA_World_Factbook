```python
from bs4 import BeautifulSoup as bs
import pandas as pd
import requests
import time
from splinter import Browser
```

```python
executable_path = {"executable_path": "chromedriver.exe"}
browser = Browser("chrome", **executable_path, headless=False)

url = 'https://www.cia.gov/library/publications/the-world-factbook/fields/208rank.html#US'
browser.visit(url)

time.sleep(5)

html = browser.html

soup = bs(html, "html.parser")
print(soup)
```

```python
tbody = soup.find("tbody")
trow = tbody.findAll("tr")
print(trow)
```

```python
country_abbr = []
country_name = []
country_gdp = []

for i in trow:
    abbr = i["id"]
    country_abbr.append(abbr)
    
    name = i.find("td", class_="region").find("a").text
    country_name.append(name)
    
    gdp = i.findAll("td")[2].text
    country_gdp.append(gdp)

print((country_abbr))
print((country_name))
print((country_gdp))
```
